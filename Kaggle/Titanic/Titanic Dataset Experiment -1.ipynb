```python
some_col["First_Name"]=some_col["Name"]
```

```python
some_col[['LastName','FirstName']] = some_col['Name'].str.split(',',expand=True)
```

```python
some_col["Designation"] = some_col['FirstName'].str.split('.',expand=True)[0]
```

```python
### Learnt how to use cut from https://github.com/pandas-dev/pandas/issues/10140
binedges = [0.0,100.0,150.0,1000]
binlabels = ["less than 100","100 to150","greater than 150"]
```

```python
some_col["Fare_Range"] = pd.cut(some_col.Fare,bins=binedges,labels=binlabels)
```

```python
some_col.groupby("Fare_Range").count()
```

```python
def fib_with_lists(n):
    fib=[0,1]
    for i in range(2,n+1):
        fib.append(fib[i-2]+fib[i-1])
    return fib[n-1]
```
